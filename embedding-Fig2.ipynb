```python
import pandas as pd
# dictionary to replace
dic = {}
cat = pd.read_csv('category2.csv')

cat.head()
cat.dropna(subset = ['Disorder'], inplace=True)
```

```python
# neoplasm -> (tumor, tumour, condition, disease)
# nos -> ''
# disease <-> disorder
for d in cat['Disorder']:
    
    phrase_list = d.lower().split(';')
    word = phrase_list[0].replace(' ', '_')
    while phrase_list:
        phrase = phrase_list.pop()
        phrase = phrase.strip()
        dic[' ' + phrase + ' '] = ' '+word+' '
        dic[' ' + phrase + 's '] = ' '+word+' '
        dic[' ' + phrase + 'es '] = ' '+word+' '
        if 'neoplasm' in phrase:
            dic[' ' + phrase.replace('neoplasm', 'tumor') + ' '] = ' '+word+' '
            dic[' ' + phrase.replace('neoplasm', 'tumour') + ' '] = ' '+word+' '
            dic[' ' + phrase.replace('neoplasm', 'condition') + ' '] = ' '+word+' '
            dic[' ' + phrase.replace('neoplasm', 'disease') + ' '] = ' '+word+' '
            dic[' ' + phrase.replace('neoplasm', 'tumors') + ' '] = ' '+word+' '
            dic[' ' + phrase.replace('neoplasm', 'tumours') + ' '] = ' '+word+' '
            dic[' ' + phrase.replace('neoplasm', 'conditions') + ' '] = ' '+word+' '
            dic[' ' + phrase.replace('neoplasm', 'diseases') + ' '] = ' '+word+' '
            location = ' '.join(phrase.split()[1:-1])
            dic[' '+'tumor of '+location+' '] = ' '+word+' '
            dic[' '+'tumors of '+location+' '] = ' '+word+' '
            dic[' '+location+'tumor'+' '] = ' '+word+' '
            dic[' '+location+'tumors'+' '] = ' '+word+' '
            dic[' '+'benign neoplasm of '+location+' '] = ' '+word+' '
            dic[' '+'benign neoplasms of '+location+' '] = ' '+word+' '  
        if 'nos ' in phrase:
            phrase_list.append(phrase.replace('nos ', ''))
        if 'disease' in phrase:
            dic[' '+phrase.replace('disease', 'disorder')+' '] = ' '+word+' '
        if 'disorder' in phrase:
            dic[' '+phrase.replace('disorder', 'disease')+' '] = ' '+word+' '
```

```python
%%time

# replace phrases in the corpus
import re
f = open('corpus/sentences_00_rev.txt')
fp = open('corpus/sentences_00_v2.txt', 'w') 

rep = dict((re.escape(k), v) for k, v in dic.items())
pattern = re.compile("|".join(rep.keys()))

#line = f.readline().lower()
while True:
    line = f.readline()
    if not line or line == '': break
    line = ' ' + re.sub(r'[^\w\s]', '', line.lower()) + ' '
    new_line = pattern.sub(lambda m: rep[re.escape(m.group(0))], line)
    fp.write(new_line)    
fp.close()
f.close()
```

# Training

```python
%%time
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
from gensim.models.word2vec import LineSentence

for corpus in ['0','1', '2']:
    for seed in [42, 0, 1]:
        sentenceCorpus = LineSentence('corpus/sentences_0%s_v2.txt'%(corpus))
        model = Word2Vec(sentences=sentenceCorpus, 
                         window=30,
                         seed=seed,
                         workers=8,)

        model.wv.save('embeddings__fig2_0%s_r%s.model'%(corpus, str(seed)))
```

```python
sentiments = pd.read_csv('sentiments.csv', index_col=0)
sentiments = sentiments[sentiments['score']<0]
sentiments['score'] *= -1
sentiments
```

```python
def summary(corpus, seed, dic, sentiments, cat):
    model = KeyedVectors.load('embeddings__fig2_0%s_r%s.model'%(corpus, seed))
    result = cat.copy()
    for senti in sentiments.index:
        result[senti] = np.nan
        for dis in cat.index:
            try:
                result.loc[dis,senti] = model.wv.similarity(dic[dis], senti)
            except:
                continue
                
    result['summary'] = 0
    for senti in sentiments.index:
        result['summary'] += sentiments.loc[senti,'score']*result[senti]
    result['summary'] /= result[sentiments.index].abs().sum(axis=1)
    return result
```
